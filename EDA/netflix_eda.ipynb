# 📊 Netflix Dataset EDA
Step-by-step analysis including pandas basics, summary statistics, correlation, trends, and conclusions.

## Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('netflix_movies_detailed_up_to_2025.csv')

df.head()
```

## Step 2: Basic Pandas Operations

```python
# Shape of dataset
df.shape

# Columns
df.columns

# Info
df.info()

# Missing values
df.isnull().sum()

# Drop duplicates
df = df.drop_duplicates()
```

## Step 3: Summary Statistics

```python
df.describe()

# Example: distribution plot
df.hist(figsize=(10,8))
plt.show()
```

## Step 4: Feature Engineering

```python
# Example: convert date column if exists
if 'release_date' in df.columns:
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
    df['year'] = df['release_date'].dt.year

df.head()
```

## Step 5: Correlation Analysis

```python
corr = df.select_dtypes(include=np.number).corr()

plt.figure()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

## Step 6: Trend Analysis

```python
# Movies vs TV Shows trend

trend_type = df.groupby(['release_year', 'type']).size().unstack()

trend_type.plot()
plt.title("Content Trend by Type")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

## Step 7: Conclusion

- Dataset cleaned and explored using pandas  
- Summary statistics reveal distribution of features  
- Correlation analysis shows relationships between numeric columns  
- Trend analysis highlights patterns over time
